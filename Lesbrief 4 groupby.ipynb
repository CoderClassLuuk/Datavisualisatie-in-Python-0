```python
import matplotlib.pyplot as plt
import pandas as pd

dfGames = pd.read_csv("games.csv")
#Alles hierboven is het standaardproces van importeren en de dataframe definen.

kolommen = ["Name", "Genre"]
#Nu de kolommen definen, ik gebruik de name voor het tellen van het aantal titels en natuurlijk gebruik ik ook de genre voor de groupby later.

dfGamesKolommen = dfGames[kolommen]
#Dan de kolommen in een var zetten

dfGamesPerGenre = dfGamesKolommen.groupby("Genre").count()
#Wat ik daarna doe, is de kolommen grouperen per genre met groupby en dan tellen hoeveel spellen het per genre zijn met de count functie.

dfGamesPerGenre.plot(kind="bar")
plt.show()
#En dan gewoon plotten voor het laatste gedeelte.
```

```python
import matplotlib.pyplot as plt
import pandas as pd

dfGames = pd.read_csv("games.csv")

kolommen = ["Name", "Year_of_Release"]

dfGamesKolommen = dfGames[kolommen]
dfGamesPerYear = dfGamesKolommen.groupby("Year_of_Release").count()

dfGamesPerYear.plot(kind="bar")
plt.show()

#Even een opmerking: de opgave vraagt om een 'gemiddeld aantal spellen per jaar', alleen kan dat niet, omdat het aantal spellen per jaar een vaste waarde is, dus wat ik heb gedaan is het aantal spellen per jaar in een grafiek zetten.
#Wat ik hier heb gedaam lijkt op de vorige opgave, alleen heb ik genre vervangen met year of release.
```
