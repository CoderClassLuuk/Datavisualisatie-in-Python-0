```python
import pandas as pd

dfGames = pd.read_csv("games.csv")

kolommenEU = ["Name", "Platform", "EU_Sales"]
kolommenNA = ["Name", "Platform", "NA_Sales"]
kolommenJP = ["Name", "Platform", "JP_Sales"]

dfGamesKolommenEU = dfGames[kolommenEU]
dfGamesNES_EU = dfGamesKolommenEU[dfGamesKolommenEU.Platform == "NES"]
dfGamesSNES_EU = dfGamesKolommenEU[dfGamesKolommenEU.Platform == "SNES"]
dfGamesWii_EU = dfGamesKolommenEU[dfGamesKolommenEU.Platform == "Wii"]
dfGamesEU = dfGamesNES_EU["EU_Sales"].sum() + dfGamesSNES_EU["EU_Sales"].sum() + dfGamesWii_EU["EU_Sales"].sum()
print(str(dfGamesEU) + " NES, SNES en Wii games verkocht in Europa.")

dfGamesKolommenNA = dfGames[kolommenNA]
dfGamesNES_NA = dfGamesKolommenNA[dfGamesKolommenNA.Platform == "NES"]
dfGamesSNES_NA = dfGamesKolommenNA[dfGamesKolommenNA.Platform == "SNES"]
dfGamesWii_NA = dfGamesKolommenNA[dfGamesKolommenNA.Platform == "Wii"]
dfGamesNA = dfGamesNES_NA["NA_Sales"].sum() + dfGamesSNES_NA["NA_Sales"].sum() + dfGamesWii_NA["NA_Sales"].sum()
print(str(dfGamesNA) + " NES, SNES en Wii games verkocht in Noord-Amerika.")

dfGamesKolommenJP = dfGames[kolommenJP]
dfGamesNES_JP = dfGamesKolommenJP[dfGamesKolommenJP.Platform == "NES"]
dfGamesSNES_JP = dfGamesKolommenJP[dfGamesKolommenJP.Platform == "SNES"]
dfGamesWii_JP = dfGamesKolommenJP[dfGamesKolommenJP.Platform == "Wii"]
dfGamesJP = dfGamesNES_JP["JP_Sales"].sum() + dfGamesSNES_JP["JP_Sales"].sum() + dfGamesWii_JP["JP_Sales"].sum()
print(str(dfGamesJP) + " NES, SNES en Wii games verkocht in Japan.")

print(str(dfGamesEU) + str(dfGamesNA) + str(dfGamesJP) + " NES, SNES en Wii games verkocht in de drie regio's bij elkaar.")
```
